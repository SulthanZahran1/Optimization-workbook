```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x = sp.symbols('x')
y = sp.symbols('y')

f = x**2 + y

constraint1 = sp.Ge(y,1)
constraint2 = sp.Gt(x,y)

df_dx = sp.diff(f,x)
df_dy = sp.diff(f,y)

```

```python
critical_points = sp.solve([df_dx, df_dy], [x,y])
critical_points
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Generate data for the plot
x_vals = np.linspace(1, 10, 100)
y_vals = np.linspace(1, 10, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2 + Y

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Plot of f(x, y) = x^2 + y')

# Add a color bar
fig.colorbar(surface, shrink=0.5, aspect=5)

plt.show()
```

Because the function is ever increasing in all directions, there is no maximum or minimum value.
