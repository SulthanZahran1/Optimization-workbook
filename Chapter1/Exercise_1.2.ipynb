```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x = sp.symbols('x')
f = x**3 - x
f

```

```python
df = sp.diff(f, x)
df
```

```python
# Find x such that df = 0
critical_points = sp.solve(df)

print("Critical points:")
for point in critical_points:
    print(f"x = {point}")

# Evaluate f at critical points
print("\nFunction values at critical points:")
function_values = []
for point in critical_points:
    value = f.subs(x, point)
    function_values.append((point, value))
    print(f"f({point}) = {value}")

# Find the minimum
minimum_point = min(function_values, key=lambda x: x[1])
print(f"\nThe minimum of the function occurs at:")
print(f"x = {minimum_point[0]}")
print(f"f(x) = {minimum_point[1]}")
```

```python
# Convert the sympy function to a numpy function using lambdify
f_numpy = sp.lambdify(x, f, 'numpy')

# Create x values
x_vals = np.linspace(-2, 2, 1000)

# Calculate y values using the numpy function
y_vals = f_numpy(x_vals)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals)
plt.scatter(minimum_point[0], minimum_point[1], color='red', label='Minimum')
plt.title("f(x) = x^3 - x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()  

# Show the plot
plt.show()

```
