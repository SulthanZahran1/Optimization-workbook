```python
import sympy as sp
sp.init_printing()
```

```python
x1 = sp.symbols('x1')
x2 = sp.symbols('x2')

f = (1-x1)**2 + 5*(x2-x1**2)**2
sp.plotting.plot_contour(f, (x1, -2, 2), (x2, -2, 2))


```

```python
f_x1 = sp.diff(f, x1)
f_x2 = sp.diff(f, x2)

f_x1_x1 = sp.diff(f_x1, x1)
f_x1_x2 = sp.diff(f_x1, x2)
f_x2_x1 = sp.diff(f_x2, x1)
f_x2_x2 = sp.diff(f_x2, x2)

ff1 = sp.Matrix([f_x1, f_x2])
ff2 = sp.Matrix([[f_x1_x1, f_x1_x2], [f_x2_x1, f_x2_x2]])

ff1.subs([(x1, 1), (x2, 1)])
```

```python
print("the gradient at (1,1) is:")
sp.pprint(ff1.subs([(x1, 1), (x2, 1)]))

print("the Hessian at (1,1) is:")
sp.pprint(ff2.subs([(x1, 1), (x2, 1)]))
```

```python
print(f"the Hessian is {ff2}")
```
